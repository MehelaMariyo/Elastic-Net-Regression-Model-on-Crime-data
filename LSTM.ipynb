```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("dataWithLags2.csv", parse_dates=True, index_col="Year")
data.head()
```

```python
data.shape
```

```python
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Example: Simulate some multivariate data
# Assume 15 features (variables) and 100 samples
data = np.random.rand(100, 26, 15)  # Shape: (samples, timesteps, features)

# Normalize the data
scaler = MinMaxScaler()
data = data.reshape(-1, 15)  # Reshape for scaling (all samples * timesteps, features)
data = scaler.fit_transform(data)
data = data.reshape(100, 26, 15)  # Reshape back to (samples, timesteps, features)

# Split into train and test sets
train_size = int(len(data) * 0.8)
train, test = data[:train_size], data[train_size:]

# Define the LSTM model
model = Sequential()
model.add(LSTM(64, input_shape=(26, 15), return_sequences=True))  # Return sequences for all timesteps
model.add(LSTM(32, return_sequences=True))  # Second LSTM also returns sequences
model.add(Dense(15))  # Output shape matches (timesteps, features)

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
history = model.fit(train, train, epochs=50, batch_size=16, validation_split=0.2, verbose=1)


# Test the model
test_loss = model.evaluate(test, test, verbose=1)
print(f"Test Loss: {test_loss}")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Predict on the test set
predictions = model.predict(test)

# Compute evaluation metrics
mse = mean_squared_error(test.reshape(-1, 15), predictions.reshape(-1, 15))
mae = mean_absolute_error(test.reshape(-1, 15), predictions.reshape(-1, 15))
r2 = r2_score(test.reshape(-1, 15), predictions.reshape(-1, 15))

print(f"MSE: {mse}, MAE: {mae}, R2: {r2}")
```

```python
from tensorflow.keras.optimizers import Adam

optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mse')
```

```python
from tensorflow.keras.layers import Dropout

model.add(Dropout(0.2))  # Add after LSTM layers
```

```python
# Predict future timesteps iteratively
future_steps = 10
input_data = test[-1]  # Use the last sequence from the test set, shape (26, 15)

predicted_future = []  # Store predictions for future timesteps

for _ in range(future_steps):
    next_step = model.predict(input_data[np.newaxis, :, :])  # Add batch dimension (1, 26, 15)
    next_step = next_step[0]  # Remove batch dimension, shape (15,)
    predicted_future.append(next_step)  # Store prediction

    # Update input_data to slide the window
    input_data = np.vstack([input_data[1:], next_step[np.newaxis, :]])  # Ensure compatible shapes

predicted_future = np.array(predicted_future)  # Convert to array for further analysis
print(predicted_future.shape)  # Shape: (10, 15)

```
