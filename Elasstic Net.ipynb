So from ACF and PACF plots we can observe that some variables should include lags. that follows ---->

## Elastic Net Regression

```python
train_size = int(len(df) * 0.8)
X_train, X_test = x[:train_size], x[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
from sklearn.linear_model import ElasticNetCV
from sklearn.linear_model import ElasticNet
# ElasticNetCV with cross-validation
elastic_net = ElasticNetCV(l1_ratio=[0.1, 0.5, 0.7, 1], alphas=[0.01, 0.1, 1, 10], cv=5)
elastic_net.fit(X_train, y_train)

# Get the best alpha and l1_ratio
print("Best alpha:", elastic_net.alpha_)
print("Best l1_ratio:", elastic_net.l1_ratio_)
```

```python
# Predict on the test set
y_pred = elastic_net.predict(X_test)

# Evaluate the model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R Score: {r2}")
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {'alpha': [0.1, 0.5, 1.0], 'l1_ratio': [0.1, 0.5, 0.9]}

# Initialize GridSearchCV
grid_search = GridSearchCV(ElasticNet(), param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the grid search to the training data
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_

# Use the best parameters to train the final model
elastic_net = ElasticNet(alpha=best_params['alpha'], l1_ratio=best_params['l1_ratio'])
elastic_net.fit(X_train, y_train)

# Predict with the tuned model
y_pred = elastic_net.predict(X_test)

# Evaluate the model
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"R-squared (Tuned): {r2}")
print(f"Mean Squared Error (Tuned): {mse}")
print(f"Root Mean Squared Error (Tuned): {rmse}")
```

```python
# Get the model coefficients
coefficients = elastic_net.coef_

# Display the coefficients with corresponding predictor names
feature_importance = pd.DataFrame({'Predictors': x.columns, 'Coefficient': coefficients})
print(feature_importance)
```

## XGBoost

```python
from xgboost import XGBRegressor
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error
# Step 4: Train an XGBoost Regressor
model = XGBRegressor(
    objective='reg:squarederror',  # Regression objective
    n_estimators=100,             # Number of trees
    learning_rate=0.1,            # Step size shrinkage
    max_depth=3                   # Max depth of trees
)

model.fit(X_train, y_train)

# Step 5: Make Predictions
y_pred = model.predict(X_test)

# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error: {rmse}")

# Optional: Feature Importance
import matplotlib.pyplot as plt
plt.barh(x.columns, model.feature_importances_)
plt.title("Feature Importance")
plt.show()
```

Future predictions

```python
def forecast_future(model, last_known_data, steps):
    """Generate future predictions iteratively."""
    future_preds = []
    input_data = last_known_data.copy()

    for _ in range(steps):
        pred = model.predict(input_data.values.reshape(1, -1))[0]
        future_preds.append(pred)

        # Shift input features and add new prediction
        input_data = input_data.shift(-1)
        input_data.iloc[-1] = pred  # Replace the last feature with the predicted value

    return np.array(future_preds)

# Get the last known row as input
last_row = x.iloc[-1, :]

# Forecast next 10 steps
future_predictions = forecast_future(model, last_row, steps=5)
print(future_predictions)
future_predictions = np.array(future_predictions)
future_predictions
```

```python
years = np.arange(2024, 2029)
years
```

```python
Future_pred = pd.DataFrame({'year': years, 'future_predictions': future_predictions})
Future_pred.set_index('year', inplace=True)
Future_pred
```

```python
import matplotlib.pyplot as plt

# Create a figure
plt.figure(figsize=(12, 6))

# Plot actual values
plt.plot(y.index, y, label="Actual", marker='o', color='blue')

# Plot predicted values
plt.plot(y_test.index, y_pred, label="Predicted", linestyle="dashed", marker='x', color='red')

plt.plot(Future_pred.index, Future_pred, label="Future", linestyle="solid", marker='x', color='black')

# Add labels and legend
plt.xlabel("Time")
plt.ylabel("Target Value")
plt.title("Actual vs Predicted Values")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

## Elastic regression for non-stationary data

```python
df = pd.read_csv('Dataa.csv')
df.head()
```

```python
yy = df['IntentionalHomicidesRate']
xx = df[['HeatIndex','UnemploymentRate','LaborForceParticipationRate','SchoolEnrollmentPrimary','GDPgrowth','InternetUsers']]
```

```python
from sklearn.model_selection import train_test_split

xx_train, xx_test, yy_train, yy_test = train_test_split(xx, yy, test_size=0.1, random_state=44)
```

```python
yy_train
```

```python
from sklearn.linear_model import ElasticNetCV
from sklearn.linear_model import ElasticNet

scaler = StandardScaler()

# Fit the scaler on the training data and transform both training and test data
xx_train = (xx_train - xx_train.mean())/xx_train.std()
xx_test = (xx_test - xx_test.mean())/xx_test.std()
yy_train = (yy_train - yy_train.mean())/yy_train.std()
yy_test = (yy_test - yy_test.mean())/yy_test.std()
# ElasticNetCV with cross-validation
elastic_net = ElasticNetCV(l1_ratio=[0.1, 0.5, 0.7, 1], alphas=[0.01, 0.1, 1, 10], cv=5)
elastic_net.fit(xx_train, yy_train)

# Get the best alpha and l1_ratio
print("Best alpha:", elastic_net.alpha_)
print("Best l1_ratio:", elastic_net.l1_ratio_)
```

```python
# Predict on the test set
yy_pred = elastic_net.predict(xx_test)

# Evaluate the model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(yy_test, yy_pred)
mae = mean_absolute_error(yy_test, yy_pred)
r2 = r2_score(yy_test, yy_pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R Score: {r2}")
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {'alpha': [0.1, 0.5, 1.0], 'l1_ratio': [0.1, 0.5, 0.9]}

# Initialize GridSearchCV
grid_search = GridSearchCV(ElasticNet(), param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the grid search to the training data
grid_search.fit(xx_train, yy_train)

# Get the best parameters
best_params = grid_search.best_params_

# Use the best parameters to train the final model
elastic_net = ElasticNet(alpha=best_params['alpha'], l1_ratio=best_params['l1_ratio'])
elastic_net.fit(xx_train, yy_train)

# Predict with the tuned model
yy_pred = elastic_net.predict(xx_test)

# Evaluate the model
r2 = r2_score(yy_test, yy_pred)
mse = mean_squared_error(yy_test, yy_pred)
rmse = np.sqrt(mse)

print(f"R-squared (Tuned): {r2}")
print(f"Mean Squared Error (Tuned): {mse}")
print(f"Root Mean Squared Error (Tuned): {rmse}")
```

```python
# Get the model coefficients
coefficients = elastic_net.coef_

# Display the coefficients with corresponding predictor names
feature_importance = pd.DataFrame({'Predictors': xx.columns, 'Coefficient': coefficients})
print(feature_importance)
```
